```python
path_var = '/content/drive/MyDrive/Colab Notebooks/data'
path_var_new = '/content/drive/MyDrive/Colab Notebooks/'
train_images = np.load(path_var_new + 'train_images_normalized.npy')
train_labels = np.load(path_var_new + 'train_labels.npy')
test_images = np.load(path_var_new + 'test_images_normalized.npy')
test_labels = np.load(path_var_new + 'test_labels.npy')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.models import Model
```

```python
# Load pretrained DenseNet201 model
densenet_model = DenseNet201(weights='imagenet', include_top=False)

# Remove top classification layer
feature_extractor = Model(inputs=densenet_model.input, outputs=densenet_model.layers[-1].output)

# Use the feature extractor to extract features from images
train_features = feature_extractor.predict(train_images)
test_features = feature_extractor.predict(test_images)

```

```python
#reshape the features into a 2-D matrix
train_features_flat = train_features.reshape(train_features.shape[0],-1)
test_features_flat = test_features.reshape(test_features.shape[0],-1)
```

```python
rom sklearn.ensemble import RandomForestClassifier
# Train Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(train_features_flat, train_labels)

# Predict labels using Random Forest classifier
rf_preds = rf_classifier.predict(test_features_flat)

# Evaluate performance
accuracy = accuracy_score(test_labels, rf_preds)
print("Random Forest Model Accuracy:", accuracy)f
```

```python
from sklearn.svm import SVC
# Train SVM classifier
svm_classifier = SVC(kernel='linear')  # You can choose different kernels (e.g., 'rbf') and tune hyperparameters
svm_classifier.fit(train_features_flat, train_labels)

# Predict labels using SVM classifier
svm_preds = svm_classifier.predict(test_features_flat)

# Evaluate performance
accuracy = accuracy_score(test_labels, svm_preds)
print("SVM Model Accuracy:", accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Standardize the features (optional but recommended)
scaler = StandardScaler()
train_features_scaled = scaler.fit_transform(train_features_flat)
test_features_scaled = scaler.transform(test_features_flat)

# Train Logistic Regression classifier
logreg_classifier = LogisticRegression(max_iter=1000, random_state=42)
logreg_classifier.fit(train_features_scaled, train_labels)

# Predict labels using Logistic Regression classifier
logreg_preds = logreg_classifier.predict(test_features_scaled)

# Evaluate performance
accuracy = accuracy_score(test_labels, logreg_preds)
print("Logistic Regression Model Accuracy:", accuracy)
```
